\begin{equation}
 x(t) = |Acos(2πF_0t)|^2

Su solucion de forma trigonometrica:


\begin{equation}
\\
f(x) = \frac{a_0}{2}∑_{n=1}^∞ a_ncos(\frac{nπx}{L}) + ∑_{n=1}^∞ b_nsen(\frac{nπx}{L})
\end{equation}

donde:

\begin{equation}
a_0 = \frac{1}{L}\int_{-L}^{L}  \! f(x) \, dx
\end{equation}

\\
\begin{equation}
a_n = \frac{1}{L}\int_{-L}^{L}  \! f(x)cos(\frac{nπx}{L}) \, dx
\end{equation}

\\
\begin{equation}
b_n = \frac{1}{L}\int_{-L}^{L}  \! f(x)sen(\frac{nπx}{L}) \, dx
\end{equation}

Definición de la señal
La señal dada es:

\begin{equation}
    x(t) = |A \cos(2\pi F_0 t)|^2
\end{equation}

Usamos la identidad trigonométrica:

\begin{equation}
    \cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}
\end{equation}

para reescribir la señal como:

\begin{equation}
    x(t) = \frac{A^2}{2} \left( 1 + \cos(4\pi F_0 t) \right)
\end{equation}

Expresión general de la serie de Fourier

La forma trigonométrica de la serie de Fourier es:

\begin{equation}
    x(t) = \frac{a_0}{2} + \sum_{n=1}^{\infty} a_n \cos \left( \frac{n\pi t}{L} \right) + \sum_{n=1}^{\infty} b_n \sin \left( \frac{n\pi t}{L} \right)
\end{equation}

donde los coeficientes se calculan como:

\begin{equation}
    a_0 = \frac{2}{T} \int_{-T/2}^{T/2} x(t) \, dt
\end{equation}

\begin{equation}
    a_n = \frac{2}{T} \int_{-T/2}^{T/2} x(t) \cos \left( \frac{n\pi t}{L} \right) dt
\end{equation}

\begin{equation}
    b_n = \frac{2}{T} \int_{-T/2}^{T/2} x(t) \sin \left( \frac{n\pi t}{L} \right) dt
\end{equation}

Dado que $ x(t) $ es una función par, todos los coeficientes $ b_n $ son cero:

\begin{equation}
    b_n = 0 \quad \forall n
\end{equation}

Cálculo de $ a_0 $

\begin{equation}
    a_0 = \frac{2}{T} \int_{-T/2}^{T/2} \frac{A^2}{2} (1 + \cos(4\pi F_0 t)) dt
\end{equation}

Separando los términos:

\begin{equation}
    a_0 = \frac{2}{T} \left( \frac{A^2}{2} \int_{-T/2}^{T/2} dt + \frac{A^2}{2} \int_{-T/2}^{T/2} \cos(4\pi F_0 t) dt \right)
\end{equation}

El primer término es:

\begin{equation}
    \frac{A^2}{2} \cdot \frac{2}{T} \cdot T = A^2
\end{equation}

El segundo término es cero, ya que la integral de un coseno en un periodo completo es cero:

\begin{equation}
    \int_{-T/2}^{T/2} \cos(4\pi F_0 t) dt = 0
\end{equation}

Por lo tanto:

\begin{equation}
    a_0 = A^2
\end{equation}

Cálculo de $ a_n $

Para $ a_n $:

\begin{equation}
    a_n = \frac{2}{T} \int_{-T/2}^{T/2} \frac{A^2}{2} (1 + \cos(4\pi F_0 t)) \cos \left( \frac{n\pi t}{L} \right) dt
\end{equation}

Nuevamente, integramos cada término por separado:

1. La integral del primer término:

   \begin{equation}
       \frac{A^2}{2} \frac{2}{T} \int_{-T/2}^{T/2} \cos \left( \frac{n\pi t}{L} \right) dt
   \end{equation}

   Esta integral es cero para todo $ n \neq 0 $, ya que la integral de un coseno en un periodo completo es cero.

2. Para el término $ \cos(4\pi F_0 t) \cos \left( \frac{n\pi t}{L} \right) $, usamos la identidad:

   \begin{equation}
       \cos(A) \cos(B) = \frac{1}{2} (\cos(A-B) + \cos(A+B))
   \end{equation}

   Aplicando esto, obtenemos términos que solo son distintos de cero cuando $ n=2 $, lo que da:

   \begin{equation}
       a_2 = \frac{A^2}{2}
   \end{equation}

   y los demás coeficientes son cero.

Por lo tanto:

\begin{equation}
    a_n =
    \begin{cases}
        A^2, & n = 0 \\
        \frac{A^2}{2}, & n = 2 \\
        0, & \text{otros valores de } n
    \end{cases}
\end{equation}

Expresión final de la serie de Fourier

Sustituyendo los coeficientes obtenidos:

\begin{equation}
    x(t) = \frac{A^2}{2} + \frac{A^2}{2} \cos(4\pi F_0 t)
\end{equation}

Esta es la serie de Fourier en forma trigonométrica.

# Exponencial

Expansión en Serie de Fourier (Forma Exponencial) para $x(t) = |A \cos(2\pi F_0 t)|^2$


Definición de la señal

La señal dada es:

\begin{equation}
    x(t) = |A \cos(2\pi F_0 t)|^2
\end{equation}

Usamos la identidad trigonométrica:

\begin{equation}
    \cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}
\end{equation}

para reescribir la señal como:

\begin{equation}
    x(t) = \frac{A^2}{2} \left( 1 + \cos(4\pi F_0 t) \right)
\end{equation}

Forma Exponencial de la Serie de Fourier

La serie de Fourier en forma exponencial está dada por:

\begin{equation}
    x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j 2\pi n F_0 t}
\end{equation}

donde los coeficientes \( C_n \) se calculan como:

\begin{equation}
    C_n = \frac{1}{T} \int_{-T/2}^{T/2} x(t) e^{-j 2\pi n F_0 t} dt
\end{equation}

Cálculo de los Coeficientes \( C_n \)

Sustituyendo \( x(t) \):

\begin{equation}
    C_n = \frac{1}{T} \int_{-T/2}^{T/2} \frac{A^2}{2} (1 + \cos(4\pi F_0 t)) e^{-j 2\pi n F_0 t} dt
\end{equation}

Descomponemos en dos términos:

\begin{equation}
    C_n = \frac{A^2}{2T} \left[ \int_{-T/2}^{T/2} e^{-j 2\pi n F_0 t} dt + \int_{-T/2}^{T/2} \cos(4\pi F_0 t) e^{-j 2\pi n F_0 t} dt \right]
\end{equation}

El primer término:

\begin{equation}
    \int_{-T/2}^{T/2} e^{-j 2\pi n F_0 t} dt
\end{equation}

Es distinto de cero solo para \( n = 0 \), en cuyo caso resulta en \( T \), por lo que:

\begin{equation}
    C_0 = \frac{A^2}{2}
\end{equation}

Para el segundo término, usamos la identidad:

\begin{equation}
    \cos(4\pi F_0 t) = \frac{e^{j 4\pi F_0 t} + e^{-j 4\pi F_0 t}}{2}
\end{equation}

Entonces:

\begin{equation}
    \int_{-T/2}^{T/2} \cos(4\pi F_0 t) e^{-j 2\pi n F_0 t} dt = \frac{1}{2} \left[ \int_{-T/2}^{T/2} e^{j 4\pi F_0 t} e^{-j 2\pi n F_0 t} dt + \int_{-T/2}^{T/2} e^{-j 4\pi F_0 t} e^{-j 2\pi n F_0 t} dt \right]
\end{equation}

Simplificando los exponentes:

\begin{equation}
    \frac{1}{2} \left[ \int_{-T/2}^{T/2} e^{-j 2\pi (n-2) F_0 t} dt + \int_{-T/2}^{T/2} e^{-j 2\pi (n+2) F_0 t} dt \right]
\end{equation}

Estas integrales son distintas de cero solo para \( n = \pm 2 \), resultando en \( T \). Entonces:

\begin{equation}
    C_{\pm 2} = \frac{A^2}{4}
\end{equation}

Expresión Final de la Serie de Fourier

Finalmente, los coeficientes son:

\begin{equation}
    C_n =
    \begin{cases}
        \frac{A^2}{2}, & n = 0 \\
        \frac{A^2}{4}, & n = \pm 2 \\
        0, & \text{otros valores de } n
    \end{cases}
\end{equation}

Por lo que la expansión en serie de Fourier en forma exponencial es:

\begin{equation}
    x(t) = \frac{A^2}{2} + \frac{A^2}{4} e^{j 4\pi F_0 t} + \frac{A^2}{4} e^{-j 4\pi F_0 t}
\end{equation}

Esta es la serie de Fourier en forma exponencial.

La transformada de Fourier de la señal modulada se puede encontrar como:

$$Y(\omega) = \mathcal{F}\{y(t)\} = \mathcal{F}\left\{\left(1+\frac{m(t)}{A_c}\right)c(t)\right\}=\mathcal{F}\{c(t)\}+\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\}$$

En este caso (utilizando tablas de Fourier):

$$C(\omega) = \mathcal{F}\{c(t)\} = \mathcal{F}\{A_c\sin(2\pi F_c t)\} = A_c\mathcal{F}\left\{\frac{e^{2\pi F_ct}-e^{-2\pi F_ct}}{2j}\right\}$$

y

$$\mathcal{F\{e^{\pm j\omega_0 t}\}} = 2\pi\delta(\omega \mp \omega_0)$$

Por consiguiente:

$$\boxed{C(\omega) = \frac{A_c\pi}{j}(\delta(\omega - 2\pi F_c)-\delta(\omega + 2\pi F_c))}$$

De forma similar:

$$\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \frac{1}{A_c}\mathcal{F}\{m(t)A_c\sin(2\pi F_c t)\} = \mathcal{F}\{m(t)\sin(2\pi F_c t)\} = \mathcal{F}\left\{\frac{m(t)e^{2\pi F_ct}-m(t)e^{-2\pi F_ct}}{2j}\right\}$$

teniendo en cuenta que:

$$\mathcal{F}\{x(t)e^{\pm j\omega_0t}\} = X(\omega \mp \omega_0)$$


Entonces:

$$\boxed{\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \frac{1}{2j}\left(M(\omega - 2\pi F_c)-M(\omega + 2\pi F_c)\right)}$$


Finalmente, el espectro de la señal modulada se puede escribir como:

```python
%%writefile pages/Modulacion.py

import streamlit as st

st.set_page_config(
    page_title="Simulación de Modulación por Amplitud (AM)",
    page_icon="🎶🎵",
)

st.title("Simulación de Modulación por Amplitud (AM)")

st.write('La modulación AM es uno de los esquemas más simples de modulación de señales y se utiliza para transmitir información (mensaje) a través de ondas portadoras de radiofrecuencia. En este ejercicio, la señal portadora es:')
st.latex(r'c(t) = A_c \sin(2\pi F_c t)')
st.write('Donde $A_c$ es la amplitud de la portadora y $F_c$ su frecuencia. El mensaje $m(t)$, que es un fragmento de audio, es modulado sobre esta portadora, generando la señal modulada:')
st.latex(r'y(t) = \left( 1 + \frac{m(t)}{A_c} \right) c(t)')
st.write('Esta señal modulada se utiliza para transmitir la información del mensaje (audio) sobre frecuencias más altas, permitiendo que la señal sea enviada de manera eficiente a largas distancias.')

st.write('La transformada de Fourier de la señal modulada se puede encontrar como:')


st.latex(r"""
Y(\omega) = \mathcal{F}\{y(t)\} = \mathcal{F}\left\{\left(1+\frac{m(t)}{A_c}\right)c(t)\right\}
= \mathcal{F}\{c(t)\} + \frac{1}{A_c}\mathcal{F}\{m(t)c(t)\}
""")

st.write("En este caso (utilizando tablas de Fourier):")

st.latex(r"""
C(\omega) = \mathcal{F}\{c(t)\} = \mathcal{F}\{A_c\sin(2\pi F_c t)\}
= A_c\mathcal{F}\left\{\frac{e^{2\pi F_c t} - e^{-2\pi F_c t}}{2j}\right\}
""")

st.write("y")

st.latex(r"""
\mathcal{F}\{e^{\pm j\omega_0 t}\} = 2\pi\delta(\omega \mp \omega_0)
""")

st.write("Por consiguiente:")

st.latex(r"""
C(\omega) = \frac{A_c\pi}{j}\left(\delta(\omega - 2\pi F_c) - \delta(\omega + 2\pi F_c)\right)
""")

st.write("De forma similar:")

st.latex(r"""
\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \frac{1}{A_c}\mathcal{F}\{m(t)A_c\sin(2\pi F_c t)\}
= \mathcal{F}\{m(t)\sin(2\pi F_c t)\} = \mathcal{F}\left\{\frac{m(t)e^{2\pi F_c t} - m(t)e^{-2\pi F_c t}}{2j}\right\}
""")

st.write("Teniendo en cuenta que:")

st.latex(r"""
\mathcal{F}\{x(t)e^{\pm j\omega_0 t}\} = X(\omega \mp \omega_0)
""")

st.write("Entonces:")

st.latex(r"""
\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \frac{1}{2j}\left(M(\omega - 2\pi F_c) - M(\omega + 2\pi F_c)\right)
""")

st.write("Finalmente, el espectro de la señal modulada se puede escribir como:")

st.latex(r"""
Y(\omega) = \frac{A_c\pi}{j}\left(\delta(\omega - 2\pi F_c) - \delta(\omega + 2\pi F_c)\right)
+ \frac{1}{2j}\left(M(\omega - 2\pi F_c) - M(\omega + 2\pi F_c)\right)
""")



```

# Distorsión Total de Armónicos (THD):
La Distorsión Total de Armónicos (THD) se refiere al grado de distorsión de una onda de corriente o tensión respecto a la onda sinusoidal ideal en un sistema eléctrico. Esta distorsión ocurre debido a la presencia de componentes armónicas, que son múltiplos enteros de la frecuencia fundamental.

El THD mide la proporción de la potencia de los armónicos respecto a la potencia de la señal fundamental, y se expresa como un porcentaje. Su valor típico se encuentra entre 0% (sin distorsión) e infinito (en casos extremos donde los armónicos dominan completamente sobre la frecuencia fundamental).

• Si ϕ=90∘, el PF es 0, lo que significa que toda la energía es reactiva y no realiza trabajo útil.

```python
%%writefile pages/Punto_3_Rectificador_Teoria.py

import streamlit as st

st.set_page_config(
    page_title="Distorsión Total de Armónicos (THD):)",
    page_icon="🧩",
)

st.title("FACTOR DE POTENCIA (PF)")

st.markdown(r"""
El factor de potencia expresa la proporción de la potencia total suministrada a un circuito que se convierte en trabajo útil, en comparación con la potencia total aparente.

**Fórmula del Factor de Potencia (PF):**

$$ PF = \frac{\text{Potencia activa}}{\text{Potencia aparente}} $$

**Donde:**
- **Potencia activa**: Es la potencia que realiza trabajo útil, medida en vatios (**W**).
- **Potencia aparente**: Es la suma de la potencia activa y reactiva, medida en voltamperios (**VA**).

El PF también puede expresarse como el coseno del ángulo de desfase (\( \phi \)) entre el voltaje y la corriente en un sistema eléctrico:

$$ PF = \cos(\phi) $$

**Donde:**
\( \phi \) = \u00c1ngulo de fase entre **I** y **V**
- Si \( \phi = 0 \) (sin desfase), el **PF** es 1, lo que indica un uso eficiente de la energía.
- Si \( \phi = 90^\circ \), el **PF** es 0, lo que significa que toda la energía es reactiva y no realiza trabajo útil.
""")

st.header("Distorsión del Factor de Potencia (DFP)")

st.markdown(r"""
El factor de potencia de distorsión es la relación entre la potencia real y la potencia aparente debido a la distorsión armónica total (**THD**).

**Fórmula del Factor de Potencia de Distorsión (PFTHD):**

$$ PF_{THD} = \frac{1}{\sqrt{1 + (THD)^2}} $$

**Donde:**
- **PFTHD** = factor de potencia de distorsión
- **THD** = distorsión armónica total

**Cálculo del THD:**

$$ THD(\%) = \left( \frac{\sqrt{P_2 + P_3 + P_4 + \cdots}}{P_1} \right) \times 100 = \left( \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1} \right) \times 100 $$

**Donde:**
- \( P_1 \) es la potencia de la frecuencia fundamental.
- \( P_2, P_3, \ldots, P_n \) es la potencia para el armónico \( n \) de la fundamental.
- Cada \( V_i \) corresponde a la amplitud respectiva del armónico.
""")
```
