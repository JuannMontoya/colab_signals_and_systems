```python
# Para graficar
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import warnings


A = 1
F_o = 5
F_s = 100
T = 1/F_s
t = np.arange(0, 1, 1/F_s)


x = A *np.cos(2*np.pi*F_o*t)
x_t = np.abs(x)**2

plt.plot(t,x_t,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("x(t)",fontsize = 14)
plt.show()
```
